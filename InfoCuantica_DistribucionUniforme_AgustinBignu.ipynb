```
import numpy as np
import matplotlib.pyplot as plt
```

```
#distribución

def distribucion(n, m):
  #hay que introducir n y m para reproducir la fórmula deseada, ejemplo: x_i+1 = ( n * x_i ) % m
  a = np.zeros(21)
  a[0] = 1
  for i in range(0,a.size-1):
    a[i+1] = (n*a[i]) % m
  
  return a

x = distribucion(13,100)
#x = np.array([1, 13, 69, 97, 61, 93, 9, 17, 21, 73, 49, 37, 81, 53, 89, 57, 41, 33, 29, 77, 1])
x = x/100
```

```
#distribución uniforme
x_prima = np.random.uniform(0,1,21)
```

```
cuenta = range(0,21)
x_prima = np.asarray(x_prima)
x_prima.size
```

```
plt.plot(cuenta, x, 'r')
plt.plot(cuenta, x_prima, 'g')

plt.show()
```

```
count, bins, ignored = plt.hist(x, 21, density=True)

plt.title('distribución dada en clase')
plt.plot(bins, np.ones_like(bins), linewidth=2, color='b')

plt.show()
```

```
count2, bins2, ignored2 = plt.hist(x_prima, 21, density=True)
plt.plot(bins2, np.ones_like(bins2), linewidth=2, color='r')
plt.title('Distribución uniforme')

plt.show()
```

```
#calculamos medias
media1 = np.mean(x)
media2 = np.mean(x_prima)

print(media1, media2)
```
